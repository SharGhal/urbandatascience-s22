```python
#All together
def woak_fare(orig):
    requestString = 'http://api.bart.gov/api/sched.aspx?cmd=fare&orig={}&dest=woak&key={}&json=y'.format(orig, APIkey)
    r = requests.get(requestString)
    d = json.loads(r.text)
    fare = d['root']['trip']['fare'] 
    return fare

print('Fare to Civic Center is ${}'.format(woak_fare('CIVC')))
print('Fare to Warm Springs/South Fremont is ${}'.format(woak_fare('WARM')))
```
